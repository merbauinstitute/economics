```python
import abcEconomics as abce
```

```python
from random import randrange
```

```python
class Customer(abce.Agent):
    def init(self, num_company):
        self.num_company = num_company
        self.create('money', 100)  
        
    def purchase(self):
        company_id = randrange(self.num_company)
        self.buy(('company', company_id), good='food', quantity=1, price=10)
    
    def print_possessions(self):
        print('    ' + self.group + str(dict(self.possessions())))
        
    
```

```python
class CompanyGoods(abce.Agent):
    def init(self):
        self.create('job', 1)
        
    def purchases(self):
        for offer in self.get_offers('food'):
            if offer.price >= 10 and self['food'] >= 1:
                self.accept(offer)
    
    def print_possessions(self):
        print('    ' + self.group + str(dict(self.possessions())))
```

```python
num_company = 1
```

```python
simulation = abce.Simulation(name='job_market', processes=1)
```

```python
companies = simulation.build_agents(CompanyGoods, 'company', number=num_company)
```

```python
customer = simulation.build_agents(Customer, 'customer', number=10, 
                                   num_company=num_company)
```

```python
food_market = companies + customer
```

```python
for r in range(100):
    simulation.advance_round(r)
    print('customer offers 10 dollar:')
    customer.purchase()
    food_market.print_possessions()
    print('company accepts or rejects the offer:')
    companies.purchases()
    food_market.print_possessions()
    print()
```
